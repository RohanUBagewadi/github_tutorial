### Itertools
This module implements a number of iterator building blocks.
Contains tool for handelling iterators.

#### Product, Permutations, Combinations, Accumulate, Groupby, and Infinite Iterrators

```python
from itertools import product
```

```python
a = [1, 2]
b = [3, 4]
prod = product(a, b) # cartician product

# The Cartesian product--- X×Y between two sets X and Y is the
# set of all possible ordered pairs with first 
# element from X and second element from Y: X×Y={(x,y):x∈X and y∈Y}.

print(prod) # creates the iterator tool
print(list(prod)) # to get the elements
```

```python
a = [1, 2]
b = [3]
prod = product(a, b, repeat=2) # cartician product
print(list(prod))
```

### 2) Permuations:
r-length tuples, all possible orderings, no repeated elements

```python
from itertools import permutations
```

```python
a = [1, 2, 3]
perm = permutations(a)
print(list(perm))
```

```python
# specifing the length of the permutation
a = [1, 2, 3]
perm = permutations(a, 2)  # returns tuples of 2 permutated
print(list(perm))
```

### 5)Accumulators: 
Make an iterator that returns accumulated sums, or accumulated results of other binary functions.

```python
from itertools import accumulate
```

```python
a = [1, 2, 3, 4]
acc = accumulate(a) # sum of elements
print(a, list(acc))
```

```python
# to multiply elements
import operator
a = [1, 2, 3, 4]
acc = accumulate(a, func=operator.mul) # returns product of elements
print(a, list(acc))
```

```python
# to multiply elements
import operator
a = [1, 2, 5, 3, 4, 9, 6]
acc = accumulate(a, func=max) # returns max comapared with each elements
print(a, list(acc))
```

#### 6) Groupby:
Makes iterator that returns keys and groups from iterable

```python
from itertools import groupby
```

```python
def smaller_then_3(x):
    return x<3

a = [1, 2, 3, 4]
group_obj = groupby(a, key=smaller_then_3)
print(group_obj)
```

```python
for key, value in group_obj:
    print(key, list(value))
```

```python
# or
a = [1, 2, 3, 4]
group_obj = groupby(a, key=lambda x: x<3)
print(group_obj)

for key, value in group_obj:
    print(key, list(value))
```

```python
persons = [{'name': 'Tim', 'age': 25}, {'name': 'Dan', 'age': 25},
          {'name': 'Max', 'age': 35}, {'name': 'Claire', 'age': 28}]

# to group people with same age
group_obj = groupby(persons, key=lambda x: x['age'])
print(group_obj)

for key, value in group_obj:
    print(key, list(value))
```

```python
from itertools import count, cycle, repeat
```

```python
# creates a infinite loop
for i in count(10): # starts from 10 to infinity
    print(i)
    
    if i>15:
        break
```

```python
a = [1, 2, 3]
# to cycle the values in list untill infinity
for i in cycle(a):
    print(i)
```

```python
a = [1, 2, 3]
# repeats the complete list specified number of times
for i in repeat(a, 4):
    print(i)
```

```python
# to repeats an integer for specified number of times
for i in repeat(2, 4):
    print(i)
```
